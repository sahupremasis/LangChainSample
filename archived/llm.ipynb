```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(
    model="gpt-4.1-mini",
    temperature=0.7,
)
```

```python
response = llm.stream("Write a poem about Driverless Car")
```

```python
for chunk in response:
    print(chunk.content, end="", flush=True)
```
